```python
import cv2
import matplotlib.pyplot as plt

gry_img= cv2.imread('lion.jpg',0)
input_image = cv2.imread('lion.jpg')

negative_image = cv2.bitwise_not(input_image)
combined_image = cv2.hconcat([input_image, negative_image])

plt.subplot(1,3,2)
plt.imshow(gry_img, cmap='gray')
plt.title('gray image'), plt.xticks([]), plt.yticks([])


plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.title('Original image')
plt.axis('off')  

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(negative_image, cv2.COLOR_BGR2RGB))
plt.title('Negative image')
plt.axis('off')  
plt.show()

```
